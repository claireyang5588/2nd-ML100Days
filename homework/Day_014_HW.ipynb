```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = 'C:/Users/claire0313/Documents/Python Scripts/Part01'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
 c=app_train.HOUSETYPE_MODE
    print(c)
```

```python
c_type=app_train.HOUSETYPE_MODE.unique()
print(c_type)
```

```python
nrows = len(c_type)
ncols = nrows // 2

plt.figure(figsize=(20,20))
for i in range(len(c_type)):
    plt.subplot(nrows, ncols, i+1)
    app_train.loc[app_train['HOUSETYPE_MODE']==c_type[i], 'AMT_CREDIT'].hist()
    plt.title(str(c_type[i]))
plt.show()
```
